```python
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]
```

```python
if mean_price > mode_price:
    print("The average price is higher than the most common price.")
elif mean_price < mode_price:
    print("The most common price is higher than the average price.")
else:
    print("The average price and the most common price are the same.")
```

```python
#Calculate the mean, median, and mode for the rating of products.
mean_rating = df['stars'].mean()
median_rating = df['stars'].median()
mode_rating = df['stars'].mode()[0]
```

```python
mean_rating
```

```python
rating_stats = df['stars'].describe()
print(rating_stats)
```

```python
#Determine the variance, standard deviation, and interquartile range for product rating.
variance_rating=df['stars'].var()
std_rating=df['stars'].std()
quantiles_rating = df['stars'].quantile([0.25, 0.5, 0.75])
```

```python
variance_rating
```

```python
std_rating
```

```python
quantiles_rating
```

```python
#Calculate the skewness and kurtosis for the rating column.
skewness_stars = df['stars'].skew()
kurtosis_stars = df['stars'].kurtosis()
```

```python
skewness_stars
```

```python
kurtosis_stars
```

```python
Positve Skewness data indicates that the data distribution is slightly skewed to the right (higher rating).
Kurtosis <3 data indicates that there are fewer outliers and the data is evenly distributed.
```
